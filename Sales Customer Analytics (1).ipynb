```python
# Apply()-row-wise

def discount_label(x):
    if x==0:
        return "Nodiscount"
    elif x<=0.10:
        return "Low Discount"
    else:
        return "High Discount"
df["DiscountCategory"]=df["DiscountPct"].apply(discount_label)
```

```python
# map()-maps the specified value to the given value

city_region={
    "Pune":"West",
    "Mumbai":"West",
    "Delhi":"North",
    "Banglore":"South",
    "Hyderabad":"South",
    "Chennai":"South"}
df["Region"]=df["City"].map(city_region)
```

```python
df["year"]=df["OrderDate"].dt.year
df["month"]=df["OrderDate"].dt.month
df["day"]=df["OrderDate"].dt.day
df["dayname"]=df["OrderDate"].dt.day_name()
df["month_name"]=df["OrderDate"].dt.month_name()
df["quarter"]=df["OrderDate"].dt.quarter
```

```python
# one hot encoding using Pandas

import pandas as pd

df = pd.DataFrame({"sales": ["CustomerName","City","Product","Category","PaymentMode","OrderStatus"]})

encoded_df = pd.get_dummies(df, columns = ["sales"], dtype=int)
print(encoded_df)
```

```python
#one hot encoding sci-kit learn
from sklearn.preprocessing import OneHotEncoder
df = pd.DataFrame({"sales": ["CustomerName","City","Product","Category","PaymentMode","OrderStatus"]})
encoder = OneHotEncoder(sparse_output = False)
encoded_fruitsdf = encoder.fit_transform(df[["sales"]])
print(encoded_fruitsdf)
```

```python
encoded_df = pd.DataFrame(encoded_fruitsdf,columns =encoder.get_feature_names_out(["fruits"]))
df1 = pd.concat([df,encoded_df],axis = 1)
print(df1)
```
